"""
该项目是为汽车购买人纳税正常情况进行数据挖掘与分析，数据来源于市场调查分析，该数据集有124个样本，14个特征向量，预测的label为输出字段，输出字段包含纳税正常和纳税异常。首先对数据进行了预处理（字符串型数据转化为数值型数据、补全缺失值、连续特征转化为离散特征），然后对数据进行了哑变量处理，哑变量处理之后得到52个特征向量，为了避免过拟合，利用PCA技术进行降维处理，分析得到40维的特征向量，将该特征向量导入到四个模型中进行比较，结果发现融合的GBDT算法的auc值最高，auc值达到了0.7708
"""

```python
data1=data.iloc[:,:-1]#特征向量
```

```python
data1.columns#查看特征向量的列名，为下步进行哑变量处理那些特征最铺垫
```

```python
data2=pd.get_dummies(data1,prefix='col_',columns=(data1.columns))#哑变量处理
```

#主成分分析PCA，由于特征向量的维度过多，避免出现过拟合现象，进一步提升模型的泛化能力。

```python
#找降到多少维合适，降维之后尽大可能地保留特征信息
pca = PCA()
pca.fit(data2)
plt.figure(1, figsize=(4, 3))
plt.clf()
plt.axes([.2, .2, .7, .7])
plt.plot(pca.explained_variance_, linewidth=2)
plt.axis('tight')
plt.xlabel('n_components')
plt.ylabel('explained_variance_')#画了一个保留方差——主成分个数趋势图
```

利用PCA发现降到40维最好，下一步将降到40维之后的数据作为新输入

```python
pca=PCA(n_components=40)
newdata=pca.fit_transform(data2)
newdata.shape
data3=pd.DataFrame(newdata)
```

```python
data3#查看链接之后的数据框
```

```python
#数据归一化
data3 = preprocessing.scale(data3)
```
